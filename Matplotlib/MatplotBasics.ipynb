```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x,y)
plt.title("Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x**2,x**2)
plt.subplot(2,2,4)
plt.plot(y**2,y**2)
```

```python
graph=plt.figure()
```

```python

fig = plt.figure(figsize=(20,8),dpi = 100)
axis1 = fig.add_axes([0.1,0.1,.4,.8]) # width,height,horizontal shift,vertical shift
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(y)
```

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
