## Pandas utility functions
#### Similarly to numpy, pandas also has a few utility functions to identify and detect null values

```python
pd.isnull(np.nan)
```

```python
pd.notnull(np.nan)
```

#### This also works for series and data frame.

```python
pd.isnull(pd.Series([1,3,None,'A']))
```

```python
pd.isnull(pd.DataFrame( #This do not tell the sum, this show how many number are null in column3.
    {
        'Column1':[1,2,3],
        'Column2':[3,4,np.nan],
        'Column3':[np.nan,5,np.nan]
    }))['Column3'].sum()
```

```python
df=pd.isnull(pd.DataFrame( 
    {
        'Column1':[1,2,3],
        'Column2':[3,4,np.nan],
        'Column3':[np.nan,5,np.nan]
    }))
df
```

## Dplicates

```python

ambassadors = pd.Series([
    'France',
    'United Kingdom',
    'United Kingdom',
    'Italy',
    'Germany',
    'Germany',
    'Germany',
], index=[
    'Gérard Araud',
    'Kim Darroch',
    'Peter Westmacott',
    'Armando Varricchio',
    'Peter Wittig',
    'Peter Ammon',
    'Klaus Scharioth '
])
```

```python
ambassadors
```

```python
ambassadors.duplicated()
```

```python
ambassadors.duplicated(keep='last')
```

```python
ambassadors.duplicated(keep=False)
```

```python
ambassadors.drop_duplicates()
```

```python
ambassadors.drop_duplicates(keep='last')
```

```python
ambassadors.drop_duplicates(keep=False)
```

```python
players = pd.DataFrame({
    'Name': [
        'Kobe Bryant',
        'LeBron James',
        'Kobe Bryant',
        'Carmelo Anthony',
        'Kobe Bryant',
    ],
    'Pos': [
        'SG',
        'SF',
        'SG',
        'SF',
        'SF'
    ]
})
```
