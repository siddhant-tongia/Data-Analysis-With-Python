# Retail Sales Analysis Using Pandas

```python
import pandas as pd

df = pd.read_csv("Sales.csv")
df
```

## 1. Data Overview

#### How many rows and columns are in the dataset?

```python
df.shape
```

What are the column names and data types?

```python
print(df.columns)
print(df.info())
```

## 2. Data Cleaning

#### Are there any missing values in the dataset?

```python
df.isnull().sum()
```

#### Are there duplicate records?

```python
df=df.drop_duplicates()
df.duplicated().sum()
```

#### Drop the row Which having the null value

```python
df=df.dropna(subset=['Postal Code'])
df.isnull().sum()
```

## 3. Sales Distribution


#### What is the total sales amount?


```python
df['Sales'].sum()
```

#### What is the average sales value per order?


```python
average_sale_value=df['Sales'].mean()
average_sale_value
```

## 4. Sales by Category


####  Which product category generates the highest sales?


```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

## 5. Sales by Region or State

#### Which region and state generates the most sales?


```python
sale_by_region=df.groupby('Region')['Sales'].sum().idxmax()
sale_by_state=df.groupby('State')['Sales'].sum().idxmax()
print(sale_by_region)
print(sale_by_state)
```

## 6. Top Selling Products

####   Which products generate the highest sales?


```python
df.groupby('Product Name')['Sales'].sum().idxmax()
```

## 7. Sales Trend Over Time


#### How do sales change over time (monthly and yearly)?


```python
df['Order Date']=pd.to_datetime(df['Order Date'],dayfirst=True)
df['Year']=df['Order Date'].dt.year
df['Month']=df['Order Date'].dt.month
print(df.groupby('Year')['Sales'].sum())
print(df.groupby('Month')['Sales'].sum())
```

## 8. Customer Segment Analysis


####  Which customer segment contributes the most sales?


```python
df.groupby('Segment')['Sales'].sum().idxmax()
```

## 9. Shipping Mode Analysis

#### Which shipping mode is used the most?

```python
df["Ship Mode"].value_counts()
```

## 10. City-Level Sales

####  Which cities generate the highest revenue?

```python
df.groupby('City')['Sales'].sum().idxmax()
```

## 11. Sales Growth Analysis

####  How does sales change from year to year?

```python
yearly_sales = df.groupby('Year')['Sales'].sum()
yearly_sales
```

## 12. Highest Sales Day

#### Which date recorded the highest total sales?


```python
df.groupby('Order Date')['Sales'].sum().idxmax()
```

## 13. Sales Forecast Idea (Simple)

####  Based on previous trends, how might sales behave next month?


```python
monthly_sales = df.groupby(['Year','Month'])['Sales'].sum()
print(monthly_sales.tail())
```

## 14. Key Business Insights


#### Region performance

```python
df.groupby('Region')['Sales'].sum()
```

#### Category should the business focus on

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

#### Products revenue

```python
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False)
```
