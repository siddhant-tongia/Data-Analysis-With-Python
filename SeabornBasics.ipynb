```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df.head()
```

Firstly while plotting different graphs we have to analyse the data and divide the columns in two groups qualitaive and quantitative.
Distribustion plot (hist,joint,pair,rug) are made of Quantitative data. In the data frame total_bill and tip are Quantitative data.
Categorical plot (box,violin,strip,swarm,bar,count) are made of Qualitative data.In the data frame sex,smoker,day,time,size are qualitative data.

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True) #kde=Kernal density estimater.
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True) #kde=Kernal density estimater.

```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter') # For comparision
```

```python
sns.pairplot(df,hue='sex') #hue divide the data into category wise.
```

```python
sns.countplot(x=df['sex'],hue=df['smoker'])
```

```python
sns.barplot(x=df['sex'],y=df['total_bill'])
```

```python
sns.barplot(x=df['sex'],y=df['total_bill'],estimator='sum')
```

```python
sns.boxplot(x=df['tip'],y=df['day'])
```

```python
sns.violinplot(x=df['day'],y=df['total_bill'])
```

```python
sns.stripplot(x='day',y='tip',data=df)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

```python
sns.swarmplot(x='tip',y='day',data=df)
sns.violinplot(x='tip',y='day',data=df)
```

Matrix plot are used to establish the relationship between the data.Matrix plot are Heat map,Cluster map,Pivot table heat map

```python
flights=sns.load_dataset('flights')
flights
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
tipscorr=tips[['total_bill','tip','size']]
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(),annot=True)
```

```python
import scipy 
sns.clustermap(tipscorr.corr())
```

```python
pvtflight=flights.pivot_table(values='passengers',index='month',columns='year')
pvtflight
```

```python
sns.heatmap(pvtflight)
```

Regeration plot used in ML

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='rainbow')
```
